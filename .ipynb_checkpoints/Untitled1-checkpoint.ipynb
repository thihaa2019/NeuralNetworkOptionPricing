```python
X = df[['Stock','Strike','Time','Volatility','Interest Rate']].values
y = df['Call Price'].values
```

```python

X_train_full, X_test, y_train_full, y_test = train_test_split(X, y, test_size=0.20, shuffle = True)
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full,shuffle = True)
X_train.shape[1:]
```

```python
X_train_full
```

# Build a 4 layer 120 nodes MLP

```python

from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, LeakyReLU, BatchNormalization

X_train.shape[1]
```

```python
nodes = 120
model = Sequential()
#MLP aka feed forward

# how many nodes should i choose
# inpuut 4 nodes (4 features)
model.add(Dense(nodes, input_dim=X_train.shape[1]))
model.add(BatchNormalization())
model.add(LeakyReLU()) # activate
model.add(Dropout(0.25)) # regularize

model.add(Dense(nodes-20, activation='elu'))
model.add(Dropout(0.25))

model.add(Dense(nodes-40, activation='relu'))
model.add(Dropout(0.25))

model.add(Dense(nodes-60, activation='elu'))
model.add(Dropout(0.25))

model.add(Dense(1,activation = "relu"))

          
model.compile(loss='mse',optimizer='adam')

```

```python
model.summary()
```

```python
History = model.fit(X_train, y_train, batch_size = 10000, epochs=50,validation_data = (X_valid,y_valid))
```

```python
history = History
history.history.keys()
```

```python

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
predict = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, predict)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, predict)
```

```python
# batch size increases overfit,
# epoch increases undefit to potimal to overfit (epoch is going back and forth once )
# batch size is number of training sample in a single batch
```

## Now we need to optimize hyperparameters(nodes, layers,learning rate,dropout)

```python

import keras
nodes = 120
def build_model(n_hidden=6, n_neurons=500, learning_rate=3e-7, input_shape=(5,),dropout = 0.3):
    model = Sequential()
    model.add(keras.layers.InputLayer(input_shape = input_shape))
    model.add(BatchNormalization())
    for layer in range(n_hidden):
        model.add(Dense(n_neurons, activation='relu'))
        model.add(Dropout(dropout))
    model.add(keras.layers.Dense(1))
    optimizer = keras.optimizers.Adam(lr = learning_rate)
    model.compile(loss='mean_squared_error', optimizer=optimizer)
    return model
```

From playing around, learning rate 3e-4 seems optimal for val loss, dropout higher than 0.20 causes overfitting.  Dropout 0.10 seems optimal.

```python
keras_reg = tf.keras.wrappers.scikit_learn.KerasRegressor(build_model)
```

```python
history = keras_reg.fit(X_train, y_train, batch_size = 10000,epochs=50,validation_data=(X_valid, y_valid))
mse_test = keras_reg.score(X_test, y_test)
y_pred = keras_reg.predict(X_test)
```

```python
# lowest parameter sertings and highest parameter setting suggest 50 epoch is enuf with 10000 batch size
pd.DataFrame(history.history).plot(figsize = (8,5))
plt.grid(True)
plt.show()
```

# Epoch and Batch Size

1. Number of observations = 3,249,000 
2. Number of (full) training observations  = 3,249,000  * 0.80 = 2,599,200
3. batch size = 10000, (325 batches, each batch consists of 10000 observations). We chose 10000 batch size to make sure a mix of all possible combinations of input is given. Generally, smaller batch optimizes better.

4. One epoch takes  325 batches (325 updates to the model) . After one epoch, we finished through the whole training set once.

5. We choose 50 epoch, hence loops through the whole training set 50 times, resulting 16,250 batches(updates) to the model.

# RandomSearch Cross Validation Hyperparameter tuning

1. Search throught the parametermeter space randomly and perform cross validation for each model. 
2. Returns the model with lowest loss error. 
3. Given we have 6 x 25 x 5 x 4 = 3,000 possible combination of parameters.
4. 1000 number of iterations search through $\frac{1000}{3000}$ = 33.33% of whole parameter space.
5. We operate with least standard number of cv 4. (i.e; take 25 percent of data to validate on each model).
6. We will have n_ter = 1000 possible models each model with 4 fold cross validation.
7. Random search saves model with smallest validation and train error. Loss function is MSE.


```python
from scipy.stats import reciprocal
from sklearn.model_selection import RandomizedSearchCV
param_distribs = {
    "n_hidden": [1,2, 3,4,5,6],
    "n_neurons": myneurons,
    "learning_rate": [3e-3,3e-4,3e-5,3e-6,3e-7],
     "dropout" : [0.0,0.10,0.20,0.30]
}
rnd_search_cv = RandomizedSearchCV(keras_reg, param_distribs, n_iter=1000,scoring = 'neg_mean_squared_error', cv=2,verbose = 1,n_jobs = 1)
history = rnd_search_cv.fit(X_train_full, y_train_full,batch_size = 10000, epochs=20)
```
