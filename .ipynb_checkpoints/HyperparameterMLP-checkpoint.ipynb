```python
import pandas as pd
spxcall = pd.read_csv("Data/spxCallDiscount.csv")
spxcall
from numpy.random import seed
seed(1)
import tensorflow
import numpy as np
tensorflow.random.set_seed(2)
import pandas as pd
import altair as alt
import tensorflow as tf

# wrong since dates are not constant 
spxcall = pd.read_csv("Data/spxCallDiscount.csv")
spxcall
spxcall['expiration'] = pd.to_datetime(spxcall.loc[:,"expiration_date"]) - pd.to_datetime(spxcall.loc[:,"data_date"])
spxcall['expiration'] = (spxcall['expiration']/np.timedelta64(1, 'D')).astype(int)
spxcall['price'] = (spxcall.loc[:,"ask_price"]+ spxcall.loc[:,"bid_price"])/2
spxcall['moneyness'] = spxcall['underlying_price']/ spxcall['strike_price']
spxcall = spxcall[["underlying_price","strike_price","expiration","moneyness","price","implied_volatility","data_date"]]
grouped = spxcall.groupby('data_date').max()
grouped['log ret'] = abs(np.log(grouped.underlying_price) - np.log(grouped.underlying_price.shift(1)))
grouped['log ret'] = grouped['log ret'] * (252)**0.5
grouped = grouped.reset_index()
grouped = grouped.loc[:,['data_date','log ret']]
spxcall = pd.concat([spxcall.set_index('data_date'),grouped.set_index('data_date')], axis=1, join='inner').reset_index()
spxcall = spxcall[spxcall['data_date']!= '2018-01-02 00:00:00'].reset_index()
spxcall = spxcall.drop(columns = 'index')
spxcall['Vol'] = spxcall['log ret']
spxcall.drop(['log ret'],axis = 1)
spxcall['Risk Free'] = np.random.uniform(1/100,3/100,size = 306675)
spxcall = spxcall[spxcall['implied_volatility']!=0].reset_index().drop(['index'],axis = 1)
spxcall['Indicator'] =0
spxcall['Indicator'][spxcall['moneyness']>1.] = 1
spxcall['Indicator'][spxcall['moneyness']<=1] = 0
```

# 1 Random search CV for IV model

```python
X = spxcall[['underlying_price','strike_price','expiration','Risk Free','Vol','moneyness','Indicator']].values
y = spxcall['price'].values
from sklearn.model_selection import train_test_split
X_train_full, X_test, y_train_full, y_test = train_test_split(X, y, test_size=0.25,random_state =1)
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full,test_size=0.25,random_state =1)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, LeakyReLU, BatchNormalization
import keras
import tensorflow as tf
nodes = 120
def build_model(n_hidden=4, n_neurons=430, learning_rate=0.0003,dropout = 0.0):
    model = Sequential()
    model.add(keras.layers.InputLayer(input_shape = (7,)))
    model.add(BatchNormalization())
    for layer in range(n_hidden):
        model.add(Dense(n_neurons, activation='relu'))
        model.add(BatchNormalization())
        model.add(Dropout(dropout))
    model.add(keras.layers.Dense(1))
    
    myoptimizer = keras.optimizers.Adam(lr = learning_rate)
        
    model.compile(loss='mean_squared_error', optimizer=myoptimizer)
    return model
keras_reg = tf.keras.wrappers.scikit_learn.KerasRegressor(build_model)
```

# RandomSearch Cross Validation Hyperparameter tuning whole space

1. Search throught the parametermeter space randomly and perform cross validation for each model. 
2. Returns the model with lowest loss error. 
3. Given we have 3840 possible combination of parameters.
4. 550 number of iterations search through $\frac{400 }{3840}$ = 10% of whole parameter space.
5. We operate with least standard number of cv 2. (i.e; take 50 percent of data to validate on each model).
6. We will have n_ter = 400  possible models each model with 2 fold cross validation.
7. Random search saves model with smallest validation and train error. Loss function is MSE.

```python
my_n_hidden = [1,2,3,4,5,6,7,8] # 1 to 8
my_n_neurons = [25,50,75,100,125,150,175,200,225,250,275,300,325,350,375,400,425,450,475,500,525,550,575,600]# 1 to 600
    
my_learning_rate = [3e-2,3e-3,3e-4,3e-5] # e-1 to e-6
my_dropout_rate = [0.00,0.1,0.2,0.3,0.4] # 0. to 0.5
(len(my_n_hidden)*len(my_n_neurons)*len(my_learning_rate)*len(my_dropout_rate))
```

```python
from sklearn.model_selection import RandomizedSearchCV
param_distribs = {
    "n_hidden": my_n_hidden,
    "n_neurons": my_n_neurons,
    "learning_rate": my_learning_rate,
    "dropout" : my_dropout_rate
}
rnd_search_cv = RandomizedSearchCV(keras_reg, param_distribs, n_iter=200,scoring = 'neg_mean_squared_error', cv=2,verbose = 1)
history = rnd_search_cv.fit(X_train_full, y_train_full,batch_size = 512, epochs=20)
```

1. Start with all possible values
2. Random search only 10% of all possible parameters space.
3. After getting the values, decrease the parameter range, increase the search space to 50%, then get results.

```python
rnd_search_cv.best_params_
```

{'n_neurons': 400, 'n_hidden': 2, 'learning_rate': 0.003, 'dropout': 0.4}

```python
rvsearch_parameters = pd.DataFrame(data=rnd_search_cv.cv_results_)
rvsearch_parameters.to_csv ('DiscountOptionsHyperparameters.csv', index = False, header=True)
#rnd_search_cv.best_params_
```
