```python

from numpy.random import seed
seed(1)
import tensorflow
import numpy as np
tensorflow.random.set_seed(2)
import pandas as pd
import altair as alt
import tensorflow as tf
call = pd.read_csv('spxCall.csv')
call['moneyness'] = call['underlying_price']/ call['strike_price']
call['Hist volatility'] =  (np.log(call.underlying_price) - np.log(call.underlying_price.shift(1))) * np.sqrt(252)
# wrong since dates are not constant 

call = call.drop(0)
call
```

```python
call
```

```python
X = call[['strike_price','underlying_price','Expiration','Hist volatility','Interest']].values
y = call['last_price'].values
from sklearn.model_selection import train_test_split
X_train_full, X_test, y_train_full, y_test = train_test_split(X, y, test_size=0.20,random_state =1)
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full,test_size=0.25,random_state =1)
X_train.shape[1:]

```

# Call option

The optimized parameters are close to parameters proposed by "Pricing options and computing implied volatilities
using neural networks"

# original best perfomrer n_hidden=6, n_neurons=500, learning_rate=3e-5, input_shape=(5,),dropout = 0.3 Following is from randomsearch

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, LeakyReLU, BatchNormalization
import keras
nodes = 120
def build_model(n_hidden=6, n_neurons=500, learning_rate=3e-5, input_shape=(5,),dropout = 0.3):
    model = Sequential()
    model.add(keras.layers.InputLayer(input_shape = input_shape))
    model.add(BatchNormalization())
    for layer in range(n_hidden):
        model.add(Dense(n_neurons, activation='relu'))
        model.add(Dropout(dropout))
    model.add(keras.layers.Dense(1))
    optimizer = keras.optimizers.Adam(lr = learning_rate)
    model.compile(loss='mean_squared_error', optimizer=optimizer)
    return model
```

```python
import keras
keras_reg = tf.keras.wrappers.scikit_learn.KerasRegressor(build_model)
history = keras_reg.fit(X_train, y_train, batch_size = 500,epochs=50,validation_data=(X_valid, y_valid),verbose = 1)
```

# Epoch vs error convergence

```python
pd.DataFrame(history.history).plot(figsize = (8,5))
import matplotlib.pyplot as plt
plt.grid(True)
plt.show()
```

```python

predict = keras_reg.predict(X_test)
from sklearn.metrics import mean_squared_error

from sklearn.metrics import r2_score
print(r2_score(y_test, predict),mean_squared_error(y_test, predict))
```

# For regression of error terms, we use L1 distance

```python
error = abs((y_test -predict.ravel()))
df = pd.DataFrame(data=X_test,columns = ['strike_price','underlying_price','Expiration','Hist volatility','Interest'] )
df['Pred Price'] =predict.ravel()
df['True Price'] = y_test
df['error'] = error
```

```python
source = df.sample(n = 500, random_state= 1,replace = False)
alt.Chart(source).mark_circle().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
).properties(
    width=150,
    height=150
).repeat(
    row=['error'],
    column=['strike_price','underlying_price', 'Expiration', 'Hist volatility','True Price','Pred Price']
).interactive()
```

# Let's do a SLR error rate vs Pred Price increase to see how much predicted price change increases error.

```python
import statsmodels.api as sm
X = df['Pred Price']
X = sm.add_constant(X)
y = df['error']
reg = sm.OLS(y,X)
results = reg.fit()
results.summary()
```

### P value suggests that this Predicted price is significant. Moreover, one unit increase in pred price increases L1 error distance by 0.05. Note this MSE is very small compared to how spread out option prices are. Hence, we have decent prediction values. Moreover, we have very good R squared value from the MLP model.

```python
import numpy as np
import scipy.stats as si
def newton_vol_call(S, K, T, C, r, sigma):
    
    #S: spot price
    #K: strike price
    #T: time to maturity
    #C: Call value
    #r: interest rate
    #sigma: volatility of underlying asset
    
    d1 = (np.log(S / K) + (r - 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = (np.log(S / K) + (r - 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    
    fx = S * si.norm.cdf(d1, 0.0, 1.0) - K * np.exp(-r * T) * si.norm.cdf(d2, 0.0, 1.0) - C
    
    vega = (1 / np.sqrt(2 * np.pi)) * S * np.sqrt(T) * np.exp(-(si.norm.cdf(d1, 0.0, 1.0) ** 2) * 0.5)
    
    tolerance = 0.000001
    x0 = sigma
    xnew  = x0
    xold = x0 - 1
        
    while abs(xnew - xold) > tolerance:
    
        xold = xnew
        xnew = (xnew - fx - C) / vega
        
        return abs(xnew)
```
