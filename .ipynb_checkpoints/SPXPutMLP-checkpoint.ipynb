# We explore SPX (European) option starting frmo 2014 Nov to 2019 Februrary

```python

import pandas as pd
import altair as alt
import tensorflow as tf
import numpy as np
put = pd.read_csv('spxPut.csv')
put['moneyness'] = put['underlying_price']/ put['strike_price']
put['Hist volatility'] =  (put.underlying_price - np.log(put.underlying_price.shift(1)))  * np.sqrt(252)


put = put.drop(0)
put
```

# Scatter plot matrix


```python
source = put.sample(n = 1000, random_state = 1,replace = False)
```

```python
source = source[['last_price','strike_price','underlying_price','Expiration','Hist volatility','moneyness']]
alt.Chart(source).mark_circle().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
).properties(
    width=150,
    height=150
).repeat(
    row=['last_price','strike_price', 'underlying_price', 'moneyness'],
    column=['Expiration', 'Hist volatility']
).interactive()
```

# Strike price and underlying price as separate predictors, we use moneyness in next model

```python
X = put[['strike_price','underlying_price','Expiration','Hist volatility']].values
y = put['last_price'].values
from sklearn.model_selection import train_test_split
X_train_full, X_test, y_train_full, y_test = train_test_split(X, y, test_size=0.20)
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full,test_size=0.25)
X_train.shape[1:]

```

# Put option

The optimized parameters are close to parameters proposed by "Pricing options and computing implied volatilities
using neural networks"

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, LeakyReLU, BatchNormalization
import keras
nodes = 120
def build_model(n_hidden=6, n_neurons=500, learning_rate=3e-5, input_shape=(4,),dropout = 0.3):
    model = Sequential()
    model.add(keras.layers.InputLayer(input_shape = input_shape))
    model.add(BatchNormalization())
    for layer in range(n_hidden):
        model.add(Dense(n_neurons, activation='relu'))
        model.add(Dropout(dropout))
    model.add(keras.layers.Dense(1))
    optimizer = keras.optimizers.Adam(lr = learning_rate)
    model.compile(loss='mean_squared_error', optimizer=optimizer)
    return model
```

```python
keras_reg = tf.keras.wrappers.scikit_learn.KerasRegressor(build_model)
history = keras_reg.fit(X_train, y_train, batch_size = 1000,epochs=100,validation_data=(X_valid, y_valid),verbose = 1)
```

# Epoch vs error convergence

```python
pd.DataFrame(history.history).plot(figsize = (8,5))
import matplotlib.pyplot as plt
plt.grid(True)
plt.show()
```

```python

predict = keras_reg.predict(X_test)
from sklearn.metrics import mean_squared_error

from sklearn.metrics import r2_score
print(r2_score(y_test, predict),mean_squared_error(y_test, predict))
```

# For regression of error terms, we use L1 distance

```python
error = (y_test -predict.ravel())
df = pd.DataFrame(data=X_test,columns = ['strike_price','underlying_price','Expiration','Hist volatility'] )
df['Pred Price'] =predict.ravel()
df['True Price'] = y_test
df['error'] = error
```

```python
source = df.sample(n = 500, random_state= 1,replace = False)
alt.Chart(source).mark_circle().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
).properties(
    width=150,
    height=150
).repeat(
    row=['error', 'Pred Price'],
    column=['strike_price','underlying_price', 'Expiration', 'Hist volatility','True Price']
).interactive()
```

### We observe that expiration has a linear trend with respect to error terms. Strike price and Underlying price seems uncertrain with respect to linear relationship with error term.
### Increase in price increases error rate 

```python
alt.Chart(source ).mark_circle().encode(
    alt.X("Pred Price", type='quantitative'),
    alt.Y("error", type='quantitative'),
).interactive()
```

# Let's do a SLR error rate vs Pred Price increase to see how much predicted price change increases error.

```python
import statsmodels.api as sm
X = df['Pred Price']
X = sm.add_constant(X)
y = df['error']
reg = sm.OLS(y,X)
results = reg.fit()
results.summary()
```

### P value suggests that this Predicted price is significant. Moreover, one unit increase in pred price increases L1 error distance by 0.07. Note this MSE is very small compared to how spread out option prices are. Hence, we have decent prediction values. Moreover, we have very good R squared value from the MLP model.

```python
print(r2_score(y_test, predict),mean_squared_error(y_test, predict))
```

```python
X  = df[['strike_price','underlying_price','Expiration','Hist volatility']]
X = sm.add_constant(X)
y = df['error']
reg = sm.OLS(y,X)
results = reg.fit()
results.summary()
```

###  We observe none of the predictors strongly influence L1 error except historical volatility. 1 unit increase in historical volatility decreases error term by 3.14. In other words, 10% increase in volatility decreases squared distance between true and predicted prices  by .314. Why does increased in volatility decreases L1 error?


## We will repeat the Neural Network but with Moneyness instead of Strike price and Spot price

```python
X = put[['moneyness','Expiration','Hist volatility']].values
y = put['last_price'].values
X_train_full, X_test, y_train_full, y_test = train_test_split(X, y, test_size=0.20)
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full,test_size=0.25)
X_train.shape[1:]
```

```python
import keras
nodes = 120
def build_model(n_hidden=6, n_neurons=500, learning_rate=3e-5, input_shape=(3,),dropout = 0.3):
    model = Sequential()
    model.add(keras.layers.InputLayer(input_shape = input_shape))
    model.add(BatchNormalization())
    for layer in range(n_hidden):
        model.add(Dense(n_neurons, activation='relu'))
        model.add(Dropout(dropout))
    model.add(keras.layers.Dense(1))
    optimizer = keras.optimizers.Adam(lr = learning_rate)
    model.compile(loss='mean_squared_error', optimizer=optimizer)
    return model
```

```python
keras_reg = tf.keras.wrappers.scikit_learn.KerasRegressor(build_model)
history = keras_reg.fit(X_train, y_train, batch_size = 1000,epochs=100,validation_data=(X_valid, y_valid))
```

```python
pd.DataFrame(history.history).plot(figsize = (8,5))
import matplotlib.pyplot as plt
plt.grid(True)
plt.show()
```

```python
import numpy as np
predict = keras_reg.predict(X_test)
mse = mean_squared_error(y_test, predict)
error =(y_test - predict.ravel())**2
r2 = r2_score(y_test, predict)
print(mse,r2)
```

# Repeat the process from moneyness as predictor

```python
error = (y_test -predict.ravel())
df = pd.DataFrame(data=X_test,columns = ['moneyness','Expiration','Hist volatility'] )
df['Pred Price'] =predict.ravel()
df['True Price'] = y_test
df['error'] = error
source = df.sample(n = 500, random_state= 1,replace = False)
alt.Chart(source).mark_circle().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
).properties(
    width=150,
    height=150
).repeat(
    row=['error', 'Pred Price'],
    column=['moneyness', 'Expiration', 'Hist volatility','True Price']
).interactive()
```

# Let's do a SLR error rate vs Pred Price increase to see how much predicted price change increases error.

```python
import statsmodels.api as sm
X = df['Pred Price']
X = sm.add_constant(X)
y = df['error']
reg = sm.OLS(y,X)
results = reg.fit()
results.summary()
```

### P value suggests that this Predicted price is significant. Moreover, one unit increase in pred price increases L1 error distance by 0.089(same as using original prediction). Note this MSE is very small compared to how spread out option prices are. Hence, we have decent prediction values. Moreover, we have very good R squared value from the MLP model.

```python
X  = df[['moneyness','Expiration','Hist volatility']]
X = sm.add_constant(X)
y = df['error']
reg = sm.OLS(y,X)
results = reg.fit()
results.summary()
```

###  Unlike the original MLP, we observe moneyness strongly influences L1 error. 1 unit increase in the moneyness (i.e; in the money), error term increases by 24.4. Moreover, 1 unit increase in historical volatility decreases error term by -2.828. Decrease is more significant. In other words, 10% increase in volatility decreases squared distance between true and predicted prices  by .2828. How does moneyness influence L1 error?
